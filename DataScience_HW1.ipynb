```python
import numpy as np
from matplotlib.pyplot import scatter
import matplotlib.pyplot as plt

# define mean vector since we are using 2 dimension my vextor would have x y
mean = [0,0]
# defining co-variance matrix
var1 = 1
var2 = 1
covs = 0
covs_matrix = [[var1,covs],[covs,var2]]
# generating x values
x = np.random.multivariate_normal(mean,covs_matrix, size = 10000)

# using scatter x[:,0] means that we are taking alll the elements of rows from 0 columns and similarly 
#the same thing for the x[:,1] , all the elements from the rows of 1 column
scatter(x[:,0], x[:,1])
plt.title("Corelated GRVs from N(0,1) Random Numbers") #title
plt.xlabel("X 1") #x label
plt.ylabel("X 2") #y label
plt.show()
```

```python
"""
part a) 

"""

import numpy as np

no_of_trials = 1000;
count_2 = 0

for i in range(no_of_trials): #iterating through individual trial
  random_integers = np.random.randint(1,11,10)  #producing array of 10 random integer
  result = np.where(random_integers == 2) # listing indexes of 2's occured in the array as "result

  # finding length of the "result" array to count occurance of 2 in every randomly gennerated array of integer between 1 - 10
  count_2 += len(result[0])


# so probablity would be
prob = count_2/(10 * no_of_trials)
print("The probablity of drawing 2 with 1000 trials is : ",prob)
```

```python
"""
part b:
"""
import matplotlib.pyplot as plt

Probability = [] #to store the probablities in the list for ( n = 10 - 1000)
trials = [] # to store the trials

for n in range(10, 1001): # range function excludes the last value so inorder to include 1000 we have to do + 1 
  count_2 = 0
  for p in range(n): # iterating through individual trial 
    random_integers = np.random.randint(1,11,10) #producing array of 10 random integer 
    result = np.where(random_integers == 2) # listing indexes of 2's occured in the array as "result" 
    count_2 += len(result[0]) # finding length of the "result" array to count occurance of 2 in every 
                                #randomly gennerated array of integer between 1 - 10

  prob = count_2/(10 * n) #calculating probablity
  Probability.append(prob) # storing probablities in the list
  trials.append(n) # storing trials in the list

#plotting graph
plt.plot(trials,Probability)
plt.xlabel("Trials")
plt.ylabel("Probability")
plt.title("Probability of the counts of outcome '2'")
plt.show()


```

## Conclusion: As we increase the value of n , the probability of drawing 2 becomes almost constant.
